```python
# Define separator for the output table
TAB = '\t'
# Define the x variable
X.<x> = GF(2)[]
# Define "Adjust to width 4" function
def adjustToWidthFour(string):
    '''
    Adjusts string's length to 
    minimum word width 4 with "0"s
    '''
    width = 4
    return "0"*(width-len(string)) + string if len(string) < width else string
```

```python
# Build field GF((2^4)^2)
# base field irreducible polynomial x^4+x+1
# extension field irreducible polynomial x^2+x+a, 
# where 'a' is an element of the base field

F.<a> = GF(2^4, modulus=x^4+x+1)
R.<y> = F.extension(x^2+x+a)

# Print table header
print("Gamma's power" + TAB + "Pairs of elements from GF(2^4)" + TAB + "Binary code" + TAB + "Polynomial")
# Print elements
for element in enumerate(R):
    idx = int(element[0]) - 1
    line = (str(idx) if idx > -1 else "-") + TAB
    right, left = element[1] / y
    line += "(" + str(left) + ", " + str(right) + ")" + TAB
    line += adjustToWidthFour(''.join(map(str,left.polynomial()))) + " "\
        + adjustToWidthFour(''.join(map(str,right.polynomial()))) + TAB
    line += str(element[1])
    print(line)
```
